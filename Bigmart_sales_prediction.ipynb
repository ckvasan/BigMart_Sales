# Checking if there are outliers and capping them, if any

```python
sns.boxplot(y='Item_Weight', data=df)
```

```python
def outlierlimit(series):
    q1 = series.quantile(0.25)
    q2 = series.quantile(0.75)
    cut_off = 1.5 * (q2 - q1)
    return (q1 - cut_off, q2 + cut_off)
```

```python
 def cappingoutliers(col):
            lower,upper = outlierlimit(df[col])
            values  = df[col].values
            df[col] =[lower if x <  lower else upper if x> upper else x for x in values ]
```

```python
sns.boxplot(y='Item_Visibility',data=df)
```

```python
cappingoutliers('Item_Visibility')
```

# Data Encoding

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=1)
```

# Data Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

# Linear Regression model 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
lr = LinearRegression()
lr.fit(scaled_X_train,y_train)
```

```python
lr_pred = lr.predict(scaled_X_test)
```

```python
print("The MAE of the Linear Regression Model", mean_absolute_error(y_test,lr_pred))
print("The R-square of the Linear Regression Model", r2_score(y_test,lr_pred))
```

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50,max_iter =100,tol=0.1)
lasso_reg.fit(scaled_X_train,y_train)
```

```python
lasso_reg.coef_
```

```python
la_pred = lasso_reg.predict(scaled_X_test)
```

```python
print("The MAE of the Lasso Regression Model", mean_absolute_error(y_test,la_pred))
print("The R-square of the Lasso Regression Model", r2_score(y_test,la_pred))
```

```python
from sklearn import linear_model

ridge_reg=linear_model.Ridge(alpha=10,max_iter=100,tol=0.1)
```

```python
ridge_pred = ridge_reg.predict(scaled_X_test)
```

```python
print("The MAE of the Ridge Regression Model", mean_absolute_error(y_test,ridge_pred))
print("The R-square of the Ridge Regression Model", r2_score(y_test,ridge_pred))
```
